```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras.applications import VGG16
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

```

```python
img_width, img_height = 224, 224

# Set the paths to the directories containing the face images
stroke_dir = 'data/stroke-and-non-acute-stroke/stroke_data'
new_stroke_dir = 'data/Strokefaces/droopy'
non_stroke_dir = 'data/stroke-and-non-acute-stroke/noStroke_data'

# Load and preprocess the images, and then display them
def load_preprocess_and_display_images(image_dir, label):
    images = []

    # Load and preprocess the images
    for filename in os.listdir(image_dir):
        img = cv2.imread(os.path.join(image_dir, filename))
        if img is not None:  # Check if the image file was successfully read
            img = cv2.resize(img, (img_width, img_height))
            images.append(img)

    # Display a grid of 9 images
    fig, axes = plt.subplots(3, 3, figsize=(10, 10))
    fig.suptitle(label)
    for i, ax in enumerate(axes.flatten()):
        if i < len(images):
            ax.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
            ax.axis('off')
        else:
            ax.axis('off')
    plt.show()

    return images

# Load, preprocess, and display images from each directory
stroke_images = load_preprocess_and_display_images(stroke_dir, 'Stroke Images')
additional_stroke_images = load_preprocess_and_display_images(new_stroke_dir, 'Additional Stroke Images')
non_stroke_images = load_preprocess_and_display_images(non_stroke_dir, 'Non-Stroke Images')

# Combine all images and corresponding labels
images = np.concatenate([stroke_images, additional_stroke_images, non_stroke_images])
labels = np.concatenate([np.ones(len(stroke_images) + len(additional_stroke_images)), np.zeros(len(non_stroke_images))])
```

```python
# Load pretrained VGG16 model without classification layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))

# Extract features from images using the VGG16 model
features = base_model.predict(images)

# Flatten the features
features = features.reshape(features.shape[0], -1)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

```

```python
# Define the Gradient Boosting Classifier model
def create_gradient_boosting_model():
    gradient_boosting = GradientBoostingClassifier()
    return gradient_boosting

# Initialize the Gradient Boosting Classifier
gradient_boosting_model = create_gradient_boosting_model()

# Define callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True, save_weights_only=False)

```

```python
# Define cross-validation strategy
cv_train = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform GridSearchCV for hyperparameter tuning
gradient_boosting_CV = GridSearchCV(estimator=gradient_boosting_model, param_grid={}, cv=cv_train, verbose=2)
gradient_boosting_CV.fit(X_train, y_train)

print("Best parameters found by GridSearchCV:")
print(gradient_boosting_CV.best_params_)

```

```python
# Make predictions on the test set using the best model
best_model = gradient_boosting_CV.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
# Evaluate the model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy}")

# Generate classification report
class_names = ["Non-Stroke", "Stroke"]
report = classification_report(y_test, y_pred, target_names=class_names)
print("Classification Report:\n", report)

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```
